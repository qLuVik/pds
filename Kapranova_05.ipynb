```python
from vega_datasets import data
cars = data.cars()

cars.head()
```

Вариант 16
Метки: mark_text(), mark_rect().

Столбцы данных: 'Horsepower', 'Miles_per_Gallon',.

Кодировки: shape, row, column.

```python

rect = alt.Chart(cars).mark_rect().encode(
    x=alt.X('Miles_per_Gallon', bin=True),
    y=alt.Y('Horsepower', bin=True),
    color='count()'
)
corl = cars[['Miles_per_Gallon','Horsepower']].corr().iloc[0,1]
text = alt.Chart(cars).mark_text(
    align="left", baseline="top"
).encode(
    x=alt.value(5),  # pixels from left
    y=alt.value(5),  # pixels from top
    text=alt.value([f"r: {corl:.3f}", 'Line 2']))


rect.interactive() + text.interactive()
```

5.2. Используя данные об автомобилях, создайте диаграмму рассеяния (*scatter-plot*), на которой *размер* (*size*) точек становится больше при наведении на них курсора.

```python
#обозначим переменную для параметров
hover = alt.selection_point(on='mouseover', nearest=True, empty=False)

chart2 = alt.Chart(cars).mark_circle().encode(
    x=alt.X('Miles_per_Gallon:Q').title(''),
    y=alt.Y('Horsepower:Q', bin=True),
    color='Origin:N',
    size=alt.condition(hover, alt.value(500), alt.value(70))
).add_params(hover)



chart2.interactive()
```

5.3. Используя данные об автомобилях, создайте двухпанельную (*two-panel*) гистограмму (количество миль на галлон на одной панели, количество лошадиных сил на другой), где вы можете перетащить мышь, чтобы выбрать данные на левой панели, чтобы отфильтровать данные на второй панели.


```python
interval = alt.selection_interval()

hist = alt.Chart(cars).mark_bar().encode(
    x='Miles_per_Gallon:Q'
    ,y='count()').add_params(interval).transform_filter(interval).interactive()

hist | hist.encode(x='Horsepower:Q')
```

5.4. Измените приведенный выше пример диаграммы разброса и гистограммы, чтобы

- панорамировать и увеличивать диаграмму рассеяния;
- гистограмма отражала только те точки, которые видны в данный момент.

```python
import pandas as pd
x = pd.DataFrame({'x': np.linspace(-len("Капранова"), len("Капранова"))})
x.head(2)
```

```python
chart = alt.Chart(x).mark_line().encode(
    x='x',
    y='sin:Q').transform_calculate(
    cos='cos(datum.x)',
    sin='sin(datum.x)').interactive()
chart + chart.mark_line(color='red').encode(y='cos:Q')
```

5.6. Используйте `transform_filter` на диаграмме задания 4 и удалите области графика, где значение кривой косинуса меньше значения кривой синуса.

```python
chart = alt.Chart(x).mark_line().encode(
    x='x',
    y='sin:Q').transform_calculate(
    cos='cos(datum.x)',
    sin='sin(datum.x)').transform_filter(
    alt.datum.cos >= alt.datum.sin
).interactive()


chart + chart.mark_line(color='red').encode(y='cos:Q')
```
