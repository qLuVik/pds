## 2.2.1: Reading and Writing Data

Файл [grades.csv](grades.csv) представляет собой файл с именами учащихся и буквенными оценками:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Записать данные из файла в массив. Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

Подсказка: функция [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) удаляет конечные пробелы из строки.

Запишите этот файл в новый файл `grades_gpa.csv`.

````python
import pandas as pd
````

````python
#Чтобы прочитать файл, сначала нужно открыть его, указав путь к файлу и указав, хотим ли прочитать (r), записать (w), и то, и другое (r+) или добавить (a).
grades_file =  open('grades.csv', 'r')
#Можно прочитать весь файл сразу.
content = grades_file.read()
content
````

````python
print(content)
````

````python
# После чтения файла требуется снова закрыть его вручную, чтобы освободить ресурсы операционной системы:
grades_file.close()
````

````python
gpas = {"A":4, "B":3, "C":2, "D":1}

data = []
with open('grades.csv', 'r') as gr:
    for line in gr:
        data.append(line.split(";"))
data
````

````python
for row in data:
    row[1] = row[1].strip()
    row.append(gpas[row[1]])

data
````

````python
with open('grades_gpa.csv', 'w') as new_file:
    for row in data:
        new_file.write(row[0] + ";"+str(row[1])+";"+str(row[2])+"\n")
````
